```python
# File path
file_path = '/kaggle/working/multispectral-object-detection/dataset/ir/ir_valid.txt'

# Absolute part you want to replace
abs_path_prefix = '/content/multispectral-object-detection/'
relative_path_prefix = 'dataset/'

# Open the file, read lines, and replace the absolute paths
with open(file_path, 'r') as file:
    lines = file.readlines()

# Replace the absolute paths with relative ones and overwrite the file
with open(file_path, 'w') as file:
    for line in lines:
        # Replace absolute path with relative path
        relative_path = line.replace(abs_path_prefix, relative_path_prefix)
        file.write(relative_path)

print(f'Relative paths have been updated in {file_path}')
```
